```python
import pandas as pd
import numpy as np
import csv
from itertools import groupby
import os, itertools
from difflib import SequenceMatcher
```

```python
df = pd.read_csv("Brands.csv")
```

```python
df
```

```python
brands_list=df["Brand"]
brands_list=brands_list.astype(str)
categories_list=df["SubCategory"]
categories_list=categories_list.astype(str)
num_orders=df["num_orders"]
```

```python
c, r = 3, 1;
brand_orders = [[None for x in range(c)] for y in range(r)]
brand_orders[0][0]=categories_list[0]
brand_orders[0][1]=brands_list[0]
brand_orders[0][2]=num_orders[0]
for i in range(1,len(brands_list)):
        for j in range(0, len(brand_orders)):
            score = SequenceMatcher(None,brands_list[i],brand_orders[j][1]).ratio()
            if(score > 0.7 and categories_list[i]==brand_orders[j][0]):
              brand_orders[j][0]=categories_list[i]
              brand_orders[j][1]=brands_list[i]
              brand_orders[j][2]=brand_orders[j][2]+num_orders[i]
            elif(score > 0.7 and categories_list[i]!=brand_orders[j][0] and j==len(brand_orders)-1):
              brand_orders.append([categories_list[i],brands_list[i],num_orders[i]])
            else:
              if(j==len(brand_orders)-1):
                brand_orders.append([categories_list[i],brands_list[i],num_orders[i]])
```
